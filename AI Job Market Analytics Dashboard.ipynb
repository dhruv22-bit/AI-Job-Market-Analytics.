JOB MARKET OVERVIEW(Charts)

Jobs by Location

```python
df["Location"].value_counts().plot(kind="bar")
sns.set_theme(palette="pastel")
plt.title("Jobs by location")
plt.show()
```

Remote vs Hybrid vs On-site

```python
df["RemoteStatus"].value_counts().plot(kind="pie",autopct="%1.1f%%")
sns.set_theme(palette="pastel")
plt.title("Work Mode Distribution")
plt.show()
```

Job Titles Distribution

```python
df["JobTitle"].value_counts().plot(kind="bar")
sns.set_theme(palette= "pastel")
plt.title("High Demand Jobs")
plt.show()
```

```python
df.groupby("JobTitle")["Salary"].mean().sort_values().plot(kind="barh")
plt.title("Average Salary by Job Title")
```

Salary Distribution

```python
sns.histplot(df["Salary"],bins=20)
plt.title("Salary Distribution")
```

Salary By Experience

```python
df.groupby("Experience")["Salary"]\
.sum()\
.plot(kind="bar")

plt.title("Average Salary by Experience")
plt.show()
```

Top Skills Demand

```python
skills = df["Skills"].str.split(", ")
all_skills = skills.explode()

all_skills.value_counts()\
.head(15)\
.plot(kind="barh")

plt.title("Most In-Demand Skills")
plt.show()
```
